```python
import pandas as pd
```

```python
npr=pd.read_csv('../UPDATED_NLP_COURSE/05-Topic-Modeling/npr.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
```

```python
dtm=tfidf.fit_transform(npr['Article'])
```

```python
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=7,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names()[2300]
```

```python
for i,t in enumerate(nmf_model.components_):
    print(f"TOP 15 WORDS FOR TOPIC #{i} ")
    print([tfidf.get_feature_names()[index] for index in t.argsort()[-15:] ])
    print('\n\n')
```

```python
topic_result=nmf_model.transform(dtm)
```

```python
nrp['Topic']= topic_result.argmax(axis=1)
```

```python

topic_dic={0:'health',1:'election',2:'legis',3:'poli',4:'random',5:'music',6:'education'}
nrp['Topic label']= nrp['Topic'].map(topic_dic)
```
